```python
df.dropna(inplace=True)
```

```python
df.columns = ['Label', 'text']
```

```python
df.head(10)
```

```python
df['Label'].value_counts()
```

```python
df['Label'].unique()
```

```python
def preprocessing_1(text):
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip().lower()
```

```python
df['text'] = df['text'].apply(preprocessing_1)
```

```python
df['Label'].value_counts()
```

```python
# There is a special structure that fast text take data:
#      __label__category text

df['Label'] = '__label__' + df['Label'].astype(str)
```

```python
df.head(5)
```

```python
df['Label'].unique()
```

```python
# we have to keep the lable in one word; fasttext desinged way
df.Label.replace('__label__Clothing & Accessories', '__label__Clothing_Accessories', inplace=True)
```

```python
df['Label'].unique()
```

```python
df['labled_texts'] = df['Label'] + ' ' + df['text']
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, stratify=df['Label'])
```

```python
train.shape, test.shape
```

```python
train.to_csv('/content/drive/MyDrive/Dataset/saved_text_fasttext/text.train', columns=['labled_texts'], index=False, header= False)
test.to_csv('/content/drive/MyDrive/Dataset/saved_text_fasttext/text.test', columns=['labled_texts'], index=False, header= False)
```

```python
import fasttext

model = fasttext.train_supervised(input='/content/drive/MyDrive/Dataset/saved_text_fasttext/text.train')

model.test('/content/drive/MyDrive/Dataset/saved_text_fasttext/text.test')
```
